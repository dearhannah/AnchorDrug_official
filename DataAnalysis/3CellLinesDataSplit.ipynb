# As Ruoqiao changed the pretrain data, so I re-splited the data.

## pretrain data

```python
cellline_map = pd.read_csv('/egr/research-aidd/menghan1/AnchorDrug/data/CellLineEncode/use_training_cell_line_expression_features_128_encoded_20240111.csv')
cell_list = cellline_map['Unnamed: 0'].to_list()
df_res = df_data[df_data['cell_iname'].isin(cell_list)] #17624 rows x 982 columns
ResDrugs = df_res.drop_duplicates(subset=['SMILES'])['SMILES'].to_list()
print(len(ResDrugs))
print(len(list(set(df_res.cell_iname.to_list()))))
df_res.to_csv(f'/egr/research-aidd/menghan1/AnchorDrug/data/HQPretrainData/resourceCelllines.csv')
print(cell_list)
df_res
```

## train & test data

```python
trainDrugs = list(set(targetDrugs).intersection(set(ResDrugs)))
print(len(trainDrugs))
testDrugs = list(set(targetDrugs)-set(ResDrugs))
print(len(testDrugs))

df_train = df_target[df_target['SMILES'].isin(trainDrugs)]
df_test = df_target[df_target['SMILES'].isin(testDrugs)]

for c in ['MCF7', 'PC3', 'A549']:
    df1 = df_train[df_train['cell_iname']==c]
    df1 = df1.groupby(by='SMILES').median(numeric_only = True).reset_index()
    # df1.to_csv(f'/egr/research-aidd/menghan1/AnchorDrug/data/newTrainData/{c}_train.csv')
    df2 = df_test[df_test['cell_iname']==c]
    df2 = df2.groupby(by='SMILES').median(numeric_only = True).reset_index()
    # df2.to_csv(f'/egr/research-aidd/menghan1/AnchorDrug/data/newTestData/{c}_test.csv')
    print(f'cell line: {c}: \ntrain data unique drugs number: {df1.shape[0]}, test data unique drugs number {df2.shape[0]}')
```

```python
# df_cell_drug_number = pd.DataFrame()
dict_cell_drug_number = {}
for c in ['MCF7', 'PC3', 'A549']+cell_list:
    df_tmp = df_data[df_data['cell_iname']==c]
    drug_tmp = df_tmp.drop_duplicates(subset=['SMILES'])['SMILES'].to_list()
    dict_cell_drug_number[c] = {'tested drug number':len(drug_tmp)}
print(dict_cell_drug_number)
```

```python
# df_cell_drug_number = 
pd.DataFrame.from_dict(dict_cell_drug_number).T.sort_values(by='tested drug number', ascending=False).to_csv('/egr/research-aidd/menghan1/AnchorDrug/DataAnalysis/drugNumberSum.csv')
```

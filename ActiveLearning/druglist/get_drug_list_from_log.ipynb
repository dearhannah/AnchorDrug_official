```python
import os
import pickle
```

```python
data_dir = '/egr/research-aidd/menghan1/AnchorDrug/ActiveLearning/logfile/druglist_100'
files = os.listdir(data_dir)
files
```

```python
for file in files:
    print(file)
    with open(f'{data_dir}/{file}', 'r') as f:
        data = f.readlines()
    idx_5 = 0
    idx_30 = 0
    for i in range(len(data)):

        if data[i]=='Round 2\n':
            print(f'Round 2: {i}')
            drug_5 = [d[:-1] for d in data[i-5:i]]
            print(drug_5)
            drug_path = f'/egr/research-aidd/menghan1/AnchorDrug/ActiveLearning/druglist/druglist_5/{file[:-4]}{idx_5}.pkl'
            with open(drug_path, 'wb') as f:
                pickle.dump(drug_5, f)
            idx_5+=1

        if data[i]=='Round 8\n':
            print(f'Round 8: {i}')
            drug_30 = [d[:-1] for d in data[i-30:i]]
            print(drug_30)
            drug_path = f'/egr/research-aidd/menghan1/AnchorDrug/ActiveLearning/druglist/druglist_30/{file[:-4]}{idx_30}.pkl'
            with open(drug_path, 'wb') as f:
                pickle.dump(drug_30, f)
            idx_30+=1
        
```
